```python
import torchvision
import torch
import torchvision.transforms as transforms
import PIL.Image as Image
import IPython.display as display
import matplotlib.pyplot as plt
```

```python
classes = [
    "Durian",
    "Pumpkin",
    "Tomato",
    "Watermelon",
]
```

```python
the_model = torch.load('C:/KPT/Model_ft_Fruit_Classification_.pt')
the_model = the_model.eval()
```

```python
mean = [0.4363, 0.4328, 0.3291]
std = [0.2129, 0.2075, 0.2038]

image_transforms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(torch.Tensor(mean), torch.Tensor(std))
])
```

```python
def classify(the_model, image_transforms, image_path, classes):
    model = the_model
    image = Image.open(image_path)
    image = image_transforms(image).float()
    image = image.unsqueeze(0)
    
    output = model(image)
    _, predicted = torch.max(output.data, 1)
    
    return(classes[predicted.item()])
```

```python
plt.figure(figsize=[20,20])
name = classify(the_model, image_transforms, 'C:/KPT/Testing/durian.jpg', classes)
img1 = Image.open('C:/KPT/Testing/durian.jpg')
plt.subplot(141);plt.imshow(img1);plt.title(name)
plt.show()

plt.figure(figsize=[20,20])
name = classify(the_model, image_transforms, 'C:/KPT/Testing/tomato.jpg', classes)
img2 = Image.open('C:/KPT/Testing/tomato.jpg')
plt.subplot(142);plt.imshow(img2);plt.title(name)
plt.show()

plt.figure(figsize=[20,20])
name = classify(the_model, image_transforms, 'C:/KPT/Testing/watermelon.jpg', classes)
img3 = Image.open('C:/KPT/Testing/watermelon.jpg')
plt.subplot(143);plt.imshow(img3);plt.title(name)
plt.show()

plt.figure(figsize=[20,20])
name = classify(the_model, image_transforms, 'C:/KPT/Testing/pumpkin.jpg', classes)
img4 = Image.open('C:/KPT/Testing/pumpkin.jpg')
plt.subplot(144);plt.imshow(img4);plt.title(name)
plt.show()
```
