```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
img = cv2.imread(path+"coins.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

def fixColor (img):
  return (cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

blurred = cv2.GaussianBlur(img, (5,5), 0)

canny = cv2.Canny(blurred, 30, 400)

plt.figure(figsize=[20,20])
plt.subplot(131);plt.imshow(fixColor(img));plt.title('Original Gray Image')
plt.subplot(132);plt.imshow(fixColor(blurred));plt.title('Blurred Image')
plt.subplot(133);plt.imshow(fixColor(canny));plt.title('Canny Image')
```

```python
(cnts, _) = cv2.findContours(canny.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cv2.drawContours(img, cnts, -1, (255, 255, 0), 10)
plt.figure(figsize=[8,8])
plt.imshow(img);plt.title('Drawn Contours')
```
